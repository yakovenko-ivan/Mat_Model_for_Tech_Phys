### 9. Локальная ошибка аппроксимации.

&nbsp; Теперь поговорим о точности разностных схем. Мы уже затрагивали этот вопрос при рассмотрении точности аппроксимации производных конечно-разностными соотношениями. Подход к определению точности конечно-разностной схемы в целом идентичен использованному нами для определения точности конечно-разностной аппроксимации производной, за тем исключением, что в конечно-разностную схему входят производные как по пространству так и по времени. Ключевой идеей при определении точности конечно-разностной схемы является её разложение в ряд Тейлора.

&nbsp; Рассмотрим схему *уголок*. Насколько хорошо расчет по этой схеме аппроксимирует решение линейного уравнения адвекции? В схеме задействованы односторонние производные $D_-(x)$ и $D_+(x)$, которые как мы знаем аппроксимируют производную в заданной точке с первым порядком точности. 
Подставим точное решение $u(x,t)$ уравнения в численную схему, при этом *локальная ошибка аппроксимации* будет определяться как:

\begin{equation}
\epsilon_i^n = \frac{u(x_i,t_{n+1}) - u(x_i,t_n)}{\Delta t} + a \frac{u(x_i,t_n) - u(x_{i-1},t_n)}{\Delta x}
\label{eq:local_truncation_error}
\end{equation}

&nbsp; Разложим точное решение в ряд Тейлора в окрестности узла сетки с координатами $(x_i,t_n)$:
\begin{equation}
u(x_i, t_{n+1}) = u(x_i,t_n) + \frac{\partial u}{\partial t} \Delta t + \frac{1}{2!} \frac{\partial^2 u}{\partial t^2} \Delta t^2 + \frac{1}{3!} \frac{\partial^3 u}{\partial t^3} \Delta t^3 + ...
\label{eq:taylor_time}
\end{equation}

\begin{equation}
u(x_{i-1}, t_n) = u(x_i,t_n) - \frac{\partial u}{\partial x} \Delta x + \frac{1}{2!} \frac{\partial^2 u}{\partial x^2} \Delta x^2 - \frac{1}{3!} \frac{\partial^3 u}{\partial x^3} \Delta x^3 + ...
\label{eq:taylor_space}
\end{equation}

&nbsp; Подставляя найденные выражения для $u(x_i, t_{n+1})$ и $u(x_{i-1}, t_n)$ в выражение для локальной ошибки аппроксимации $\eqref{eq:local_truncation_error}$ получим следующую форму записи этой величины:

\begin{equation}
\epsilon_i^n = \frac{u(x_i,t_{n+1}) - u(x_i,t_n)}{\Delta t} + a \frac{u(x_i,t_n) - u(x_{i-1},t_n)}{\Delta x} = \underbrace{\left( \frac{\partial u}{\partial t} + a \frac{\partial u}{\partial x} \right)}_\textrm{ = 0} + G(\Delta x, \Delta t)
\label{eq:local_truncation_error2}
\end{equation}

где 
\begin{equation}
G(\Delta x, \Delta t) = \left( \frac{1}{2!} \frac{\partial^2 u}{\partial t^2} \Delta t - a \frac{1}{2!} \frac{\partial^2 u}{\partial x^2} \Delta x \right) + \left( \frac{1}{3!} \frac{\partial^3 u}{\partial t^3} \Delta t^2 + a \frac{1}{3!} \frac{\partial^3 u}{\partial x^3} \Delta x^2 \right) + ... + \left( \frac{1}{k!} \frac{\partial^k u}{\partial t^k} \Delta t^{k-1} - (-1)^k a \frac{1}{k!} \frac{\partial^k u}{\partial x^k} \Delta x^{k-1} \right) + ...
\label{eq:differential_approximation}
\end{equation}

&nbsp; Функция $G(\Delta x, \Delta t)$ называется *дифференциальным приближением* разностной схемы. Производные времени, входящие в $G(\Delta x, \Delta t)$ удобно выразить через пространственные производные:
\begin{equation*}
\frac{\partial^k u}{\partial t^k} = (- a)^k \frac{\partial^k u}{\partial x^k} 
\end{equation*}

Тогда дифференциальное приближение $G(\Delta x, \Delta t)$ примет следующую форму, которую мы будем называть *P - формой дифференциального приближения*:

\begin{equation}
G(\Delta x, \Delta t) = P(\Delta x, \Delta t) =  \frac{c}{2!} \frac{\partial^2 u}{\partial x^2} (a \Delta t - \Delta x) + (-1)^3 \frac{c}{3!} \frac{\partial^3 u}{\partial x^3} (a^2 \Delta t^2 - \Delta x^2) + ... (-1)^k \frac{c}{k!} \frac{\partial^k u}{\partial x^k} (a^{k-1} \Delta t^{k-1} - \Delta x^{k-1}) + ...
\label{eq:P_differential_approximation}
\end{equation}

&nbsp; Члены P - формы дифференциального приближения $\eqref{eq:P_differential_approximation}$ с наименьшими степенями $\Delta x$ и $\Delta t$ называют *первым дифференциальным приближением*. Степень $\Delta x$ и $\Delta t$ в первом дифференциальном приближении является *порядком точности* конечно-разностной схемы. Таким образом, для схемы уголок первое дифференциальное приближение имеет вид:

\begin{equation}
P(\Delta x, \Delta t) =  \frac{c}{2!} \frac{\partial^2 u}{\partial x^2} (a \Delta t - \Delta x)
\label{eq:SU_differential_approximation}
\end{equation}

&nbsp; Откуда можно видеть, что схема имеет первый порядок точности по времени и пространству.
Локальная ошибка аппроксимации для схемы:

\begin{equation}
\epsilon_i^n = \frac{u(x_i,t_{n+1}) - u(x_i,t_n)}{\Delta t} + a \frac{u(x_i,t_n) - u(x_{i-1},t_n)}{\Delta x} = \\ \underbrace{\left( \frac{\partial u}{\partial t} + a \frac{\partial u}{\partial x} \right)}_\textrm{ = 0} +  \frac{c}{2!} \frac{\partial^2 u}{\partial x^2} (a \Delta t - \Delta x) = O(\Delta t) + O(\Delta x)
\label{eq:local_truncation_error3}
\end{equation}

&nbsp; Для схемы с симметричной аппроксимацией производной по пространству FTCS нам необходимо также выражение для $u(x_{i+1},t_n)$:

\begin{equation}
u(x_{i+1}, t_n) = u(x_i,t_n) + \frac{\partial u}{\partial x} \Delta x + \frac{1}{2!} \frac{\partial^2 u}{\partial x^2} \Delta x^2 + \frac{1}{3!} \frac{\partial^3 u}{\partial x^3} \Delta x^3 + ...
\label{eq:taylor_space_forward}
\end{equation}

&nbsp; Используя разложения $\eqref{eq:taylor_space}$, $\eqref{eq:taylor_time}$ и $\eqref{eq:taylor_space_forward}$ получим первое дифференциальное приближение:

\begin{equation}
P(\Delta x, \Delta t) =  \frac{c^2}{2!} \frac{\partial^2 u}{\partial x^2} \Delta t + \frac{c}{3!} \frac{\partial^3 u}{\partial x^3} \Delta x^2
\label{eq:FTCS_differential_approximation}
\end{equation}

&nbsp; Таким образом, схема FTCS имеет первый порядок точности по времени и второй по пространству.

&nbsp; Наконец, используя дополнительно разложения для $u(x_n, t_{n-1})$ и $u(x_{n+1}, t_{n+1})$: 

\begin{equation}
u(x_i, t_{n-1}) = u(x_i,t_n) - \frac{\partial u}{\partial t} \Delta t + \frac{1}{2!} \frac{\partial^2 u}{\partial t^2} \Delta t^2 - \frac{1}{3!} \frac{\partial^3 u}{\partial t^3} \Delta t^3 + ...
\label{eq:taylor_time_backward}
\end{equation}

\begin{equation}
u(x_{i+1}, t_{n+1}) = u(x_i,t_n) + \frac{\partial u}{\partial t} \Delta t + \frac{\partial u}{\partial x} \Delta x + \left(\frac{1}{2!} \frac{\partial^2 u}{\partial t^2} \Delta t^2 + \frac{\partial^2 u}{\partial x \partial t} \Delta t \Delta x + \frac{1}{2!} \frac{\partial^2 u}{\partial x^2} \Delta x^2 \right)  \\ + \left( \frac{1}{3!} \frac{\partial^3 u}{\partial t^3} \Delta t^3 + \frac{1}{2!} \frac{\partial^3 u}{\partial x^2 \partial t} \Delta x^2 \Delta t + \frac{1}{2!} \frac{\partial^3 u}{\partial x \partial t^2} \Delta x \Delta t^2 + \frac{1}{3!} \frac{\partial^3 u}{\partial x^3} \Delta x^3 \right) + ...
\label{eq:taylor_diag}
\end{equation}

мы можем найти дифференциальные приближения для всех остальных схем, которые были нами рассмотрены ранее.

### 10. Свойства дифференциального приближения.

&nbsp; Из анализа процедуры построение дифференциального приближения можно придти к выводу о том, что разностная схема с большей точностью аппроксимирует свое дифференциальное приближение, чем исходное уравнение переноса. Проанализировав решение дифференциального приближения мы можем предсказать поведение решения, полученного по разностной схеме. В общем виде дифференциальные приближения разных схем для уравнения переноса выглядят следующим образом:

\begin{equation}
\frac{\partial u}{\partial t} + a \frac{\partial u}{\partial x} + \mu \frac{\partial^2 u}{\partial x^2} + \eta \frac{\partial^3 u}{\partial x^3} = 0
\label{eq:differential_approximation_general}
\end{equation}

&nbsp; Частное решение данного уравнения можно искать в виде бегущей волны.

\begin{equation}
u(x,t) = A e^{i(\omega t- kx)}
\label{eq:wave_solution}
\end{equation}

&nbsp; Подставляя данное решение в $\eqref{eq:differential_approximation_general}$ получим к уравнению, связывающему круговую частоту $\omega$ с волновым вектором $k$:

\begin{equation}
\omega - ak -i \mu k^2 + \eta k^3 = 0
\label{eq:dispersion_equation}
\end{equation}

&nbsp; Уравнение $\eqref{eq:dispersion_equation}$ называется *дисперсионным соотношением*. Выражая $\omega$ через $k$ и подставляя это выражение в $\eqref{eq:wave_solution}$, получим частное решение:

\begin{equation}
u(x,t) = A e^{\mu k^2 t} e^{-ik [x - (a-\eta k^2)t]}
\label{eq:wave_solution2}
\end{equation}

&nbsp; При $\mu < 0$ наблюдается *диссипация*, амплитуда волны  $\eqref{eq:wave_solution2}$ будет экспоненциально затухать, причем короткие волны будут затухать быстрее. Если $\mu > 0$, то наоборот, будет наблюдаться рост амплитуды гармоник.

&nbsp; Величина 
\begin{equation}
\gamma (k) = \frac{Re[\omega]}{ak}
\label{eq:phase_velo}
\end{equation}

называется *приведенной фазовой скоростью* бегущей волны. Зависимость $\gamma$ от $k$ определяет *дисперсию* - волны с разными длинами распространяются с различной скоростью. Если $\gamma (k) < 1$, дисперсию называют нормальной, в обратном случае, при $\gamma (k) > 1$ дисперсия является аномальной.

### 11. Глобальная ошибка.

&nbsp; Итак, мы научились находить локальную ошибку аппроксимации в точке, возникающую при использовании разностных схем. Но как эта локальная ошибка связана с ошибкой решения во всей расчетной области? В первую очередь необходимо определить что такое *глобальная ошибка* решения и способы её измерения. Пусть вектор значений истинного решения в узлах сетки на временном шаге $t_n$ записывается как:

\begin{equation*}
\vec u(x,t_n) = 
 \begin{pmatrix}
  u(x_0,t_n) \\
  u(x_1,t_n) \\
  \vdots    \\
  u(x_N,t_n) 
 \end{pmatrix}
\end{equation*}

&nbsp; В свою очередь приближенное решение в узлах сетки определяется вектором:

\begin{equation*}
\vec u^n = 
 \begin{pmatrix}
  u_0^n \\
  u_1^n \\
  \vdots    \\
  u_N^n 
 \end{pmatrix}
\end{equation*}

&nbsp; Вектор состоящий из погрешность приближения точного решения численным решением в каждом узле определяется как разность между векторами $\vec u(x,t_n)$ и $\vec u^n$:

\begin{equation}
\vec E = \vec u^n - \vec u(x,t_n)
\label{eq:global_error}
\end{equation}

&nbsp; Для того, чтобы сопоставить вектору ошибки определенное значение можно использовать разные варианты нормы, относящиеся как правило к семейству Гёльдеровых норм $$\|x\|_p = \left( \sum_i {|x_i|^p} \right)^\frac{1}{p}$$

&nbsp; К примеру для $p \to \infty$ получим max-норму:

\begin{equation}
\| \vec E \|_\infty = \max_{0 \le i \le N} |E_i| = \max_{0 \le i \le N} | u^n_i - u(x_i,t_n)|
\label{eq:max_norm_global_error}
\end{equation}

&nbsp; В данном случае глобальная ошибка определяется просто как наибольшая ошибка в расчетной области на заданном временном слое. Если нам удается показать, что $\| \vec E \|_\infty = O(h^2)$ это автоматически гаранирует, что ошибка в любой точке области также $O(h^2)$.

&nbsp; Широко используются также нормы $p=1$ и $p=2$:

\begin{equation}
\| \vec E \|_1 = \Delta x \sum_{i=1}^N |E_i| 
\label{eq:1_norm_global_error}
\end{equation}

\begin{equation}
\| \vec E \|_2 = \left( \Delta x \sum_{i=1}^N |E_i|^2 \right)^\frac{1}{2} 
\label{eq:2_norm_global_error}
\end{equation}

&nbsp; Не сложно проследить связь между локальной ошибкой аппроксимации $\epsilon_i^n$ и глобальной ошибкой $\vec E$. Определим вектор $\vec \epsilon$ с элементами, равными локальной ошибке аппроксимации в узлах сетки:

\begin{equation}
\vec \epsilon = \mathbf{A} \vec u(x,t_n)
\label{eq:LTE_vec}
\end{equation}

где $\mathbf{A}$ - матричное представление конечно-разностного оператора. Вычитая из $\eqref{eq:LTE_vec}$ $\mathbf{A} \vec u^n = 0$ (конечно-разностное представление дифференциального уравнения для сеточной функции), получим:

\begin{equation}
-\vec \epsilon = \mathbf{A} \vec u^n - \mathbf{A} \vec u(x,t_n) = \mathbf{A} \vec E
\label{eq:LTE_GE}
\end{equation}

&nbsp; К примеру, для схемы уголок: 

\begin{equation}
-\vec \epsilon = \frac{E^{n+1}_i - E^n_i}{\Delta t} + a \frac{E^n_i-E^n_{i-1}}{\Delta x}
\label{eq:simple_upwind_forward_GE}
\end{equation}

### 12. Согласованность, устойчивость и сходимость численной схемы.

&nbsp; Схема называется ***согласованной*** с дифференциальным уравнением и граничными условиями, если 

\begin{equation}
\|\vec\epsilon_{\Delta x, \Delta t}\| = 
\max_{i,n}|\epsilon_i^n| \to  0 \: при \: \Delta x \to 0 \: и \: \Delta t \to 0
\label{eq:consistency}
\end{equation}

&nbsp; Здесь индекс $\Delta x, \Delta t$ означает расчет на сетке с ячейкой $\Delta x, \Delta t$. Проще говоря локальная ошибка аппроксимации в любой точке расчетной области должна уменьшаться при уменьшении шагов сетки $\Delta x$ и $\Delta t$. Таким образом, для согласованности схемы необходимо, чтобы схема была по крайней мере первого порядка точности по времени и пространству $\|\vec\epsilon_{\Delta x, \Delta t}\| = O(\Delta x^p,\Delta t^m)$, $p>1$, $m>1$.

&nbsp; Схема является ***сходящейся***, если глобальная ошибка стремится к нулю по мере измельчения сетки 

\begin{equation}
\|\vec E_{\Delta x, \Delta t}\| = 
\max_{i,n}|E_i^n| \to  0 \: при \: \Delta x \to 0 \: и \: \Delta t \to 0
\label{eq:convergence}
\end{equation}

&nbsp; Выше мы уже нашли выражения для локальной ошибки аппроксимаций некоторых схем, из которых можно заключить, что все они являются согласованными. Однако, схема может быть согласованной, но при этом не быть сходящейся. Продемонстрируем это на простом примере.

&nbsp; Рассмотрим схему *уголок*. Её локальная ошибка аппроксимации $\eqref{eq:local_truncation_error2}$ стремится к нулю по мере измельчения сетки ($\Delta x$ и $\Delta t \to 0$) и следовательно это согласованная схема. Давайте проанализируем численное решение уравнения адвекции с наклоном характеристики как показано на [рисунке 6](#figure6). Характеристика, проходящая через точку $A$, проходит также через точку $B$ на оси $0x$. Следовательно, точное решение в $A$ равно начальному значению в точке $B$. Однако численное решение в точке $A$ вычисляется с помощью значений в точках внутри серой области. Данная область называется *областью зависимости* численной схемы. Точка $B$ лежит вне этой области и поэтому не может влиять на решение в точке $A$. Следовательно, ошибка в точке $A$ может быть произвольно большой. Если мы будем пропроционально уменьшать пространственный и временной шаги то область зависимости останется той же самой и ситуация не изменится. Таким образом, пока отношение шагов $\Delta x$ и $\Delta t$ остается постоянным, уменьшение шагов сетки не позволит уменьшить глобальную ошибку численного решения. 
&nbsp; Очевидно, что для сходимости схемы *уголок* необходимо, чтобы характеристика, определяющая точное решение в любой точке сетки, находилась внутри области зависимости численного решения для этой точки. Это произойдет, когда наклон характеристик будет больше, чем наклон границы области зависимости, т.е. когда:

\begin{equation}
a\Delta t \le \Delta x \rightarrow C=\frac{a \Delta t}{\Delta x} \le 1
\label{eq:necessary_convergence}
\end{equation}

&nbsp; Таким образом, $\eqref{eq:necessary_convergence}$ есть необходимое условие сходимости схемы *уголок*.

&nbsp; Если известно, что точное решение дифференциального уравнения ограничено, для ***устойчивости*** схемы необходимо, чтобы численное решение также оставалась ограниченным с увеличением временного слоя $n$ при фиксированных значениях $\Delta x$ и $\Delta t$:

\begin{equation}
\|\vec u^n_{\Delta x, \Delta t}\| \le \|\vec u^0_{\Delta x, \Delta t}\| \: для \: любого \: n \Delta t \le T 
\label{eq:stability}
\end{equation}

где $T$ - произвольный момент времени.

### 13. Спектральный анализ разностных схем.

&nbsp; Как следует из теоремы Лакса, для доказательства сходимости метода достаточно показать, что он согласован и устойчив. Если проверка согласованности как правило не составляет труда, то с устойчивостью дело обстоит куда сложнее. Существует большое разнообразие различных методов аналитического доказательства устойчивости конечно-разностных схем, к примеру обосновать устойчивость в max-норме можно с помощью так называемого принципа максимума, устойчивость в норме $p=2$ можно показать используя методы энергетических неравенств, методов операторных неравенств, разделения переменных и замораживания коэффициентов. Все эти методы обладают разными границами применимости и трудоемкостью. Подробное описание различных подходов к обоснованию устойчивости разностных схем выходит за пределы настоящего курса, для более подробного ознакомления с методами аналитического доказательства устойчивости схем рекомендуются монографии [[4]](#foot4)[[5]](#foot5)[[6]](#foot6). В монографии приведен обзор и оценка наиболее популярных методов исследования устойчивости разностных схем, используемых при решении задач гидродинамики [[7]](#foot7).

&nbsp; Из всех методов анализа устойчивости разностных схем одним из наиболее универсальных и простых является спектральный анализ, предложенный Джоном фон Нейманом [[8]](#foot8). Метод Неймана позволяет исследовать не только устойчивость разностных схем, но и другие важные свойства, такие как численная диссипация и дисперсия. 

&nbsp; Метод Неймана заключается в представлении решения в форме ряда Фурье и анализе поведения отдельных гармоник при переходе с одного временного слоя на другой. В результате, устойчивость всех гармоник будет необходимым условием устойчивости схемы. 

&nbsp; Представим сеточное решение в виде бегущих волн:

\begin{equation}
u^n_j = A e^{i(\omega n \Delta t- k j \Delta x)} = A q^n e^{-i( k j \Delta x)}
\label{eq:wave_solution_FD}
\end{equation}

здесь индекс $j$ введен как пространственный индекс ячейки, чтобы не путать его с мнимой единицей $i$. $q$ называется *множителем перехода*, $|q|$ определяет будет ли решение возрастать или убывать при переходе с временного слоя $n$ на временной слой $n+1$.

&nbsp; После подстановки в конечно-разностное уравнение схемы *уголок* получим:
\begin{equation*}
\frac{A q^{n+1}e^{-ikj\Delta x} - A q^{n}e^{-ikj\Delta x}}{\Delta t} = - a \frac{A q^{n}e^{-ikj\Delta x}-A q^{n}e^{-ik(j-1)\Delta x}}{\Delta x}
\end{equation*}

&nbsp; Разделив на $A e^{ikj\Delta x}$ и умножив на $\Delta t$ придем к следующему уравнению:

\begin{equation*}
q^{n+1} = (1-С) q^{n} + C q^{n}e^{ik\Delta x}
\end{equation*}

&nbsp; Осюда имеем 
\begin{equation}
q = (1-С) + C e^{ik\Delta x}
\label{eq:q_upwind}
\end{equation}

&nbsp; Уравнение $\eqref{eq:q_upwind}$, связывающее множитель перехода с параметрами конечно-разностной схемы называется *характеристическим уравнением* разностной схемы.

&nbsp; Вычисляя $|q|$, получим:

\begin{equation}
|q(C,k\Delta x)| = \sqrt{[(1-C)+C \cos (kh)]^2 + [C \sin (kh)]^2}
\label{eq:absq_upwind}
\end{equation}

&nbsp; Модуль множителя перехода является функцией от числа $C$ и *приведенного волнового числа* $kh$, изменяющегося в пределах от $-\pi$ до $\pi$. Если модуль множителя перехода конечно-разностной схемы меньше или равен 1, то амплитуда волны не возрастает. 


#### Необходимое условие устойчивости разностной схемы по фон Нейману.
> Говорят, что конечно разностная схема при заданном числе Куранта-Фридрихса-Леви $C$ является устойчивой, если амплитуды *всех* волн $k\Delta x \in (-\pi, \pi) $ не возрастают, т.е. $|q| \le 1$ для любого $k\Delta x \in (-\pi, \pi) $. 
 
Из $\eqref{eq:q_upwind}$ следует также дисперсионное соотношение:

\begin{equation*}
q = e^{i \omega \Delta t} = (1-С) + C e^{ik\Delta x}
\end{equation*}

Откуда

\begin{equation}
\omega = -\frac{i}{\Delta t} \arg\left[(1-С) + C e^{ik\Delta x}\right]
\end{equation}

Подставляя в $\eqref{eq:phase_velo}$, получим выражение для приведенной фазовой скорости:

\begin{equation*}
\gamma(C,k\Delta x)=\frac{\omega}{ka} = -\frac{i}{\Delta t k a} \arg\left[(1-С) + C e^{ik\Delta x}\right] = -\frac{i}{C k \Delta x} \arg\left[(1-С) + C e^{ik\Delta x}\right]
\end{equation*}

&nbsp; Проанализируем спектральные свойства разностной схемы уголок, используя возможности символьных вычислений **SymPy**. Для этого, найдем запишем $\eqref{eq:q_upwind}$ в символьном виде, и с помощью процедур [`Abs`](https://docs.sympy.org/latest/modules/functions/elementary.html) и [`arg`](https://docs.sympy.org/latest/modules/functions/elementary.html) найдем $|q(C,k\Delta x)|$ и $|\gamma(C,k\Delta x)|$, и преобразуем их сразу в python функции:

```python
from numpy import pi
from sympy import exp, sin, cos, sqrt
from sympy import symbols, lambdify
from sympy import I, Abs, arg

from sympy.solvers.inequalities import solve_univariate_inequality

q, abs_q, C, k, dx = symbols("q, abs_q, C,k,dx", real=True)

q = C*exp(I*k*dx)+(1-C)                                      # q(C,k dx)            
abs_q = (Abs(q)).rewrite(cos).simplify().collect(cos(dx*k))  # |q(C,k dx)| 
arg_q = Abs(-I/k/dx/C*arg(q))                                # |1/(Ckdx) * arg(q(C,k dx))|
q_func_upwind = lambdify([C,dx,k],abs_q)                     # python функция, рассчитывающая |q(C,k dx)| 
g_func_upwind = lambdify([C,dx,k],arg_q)                     # python функция, рассчитывающая |gamma(C,k dx)| 
```

&nbsp; Для того, чтобы определить, при каких значениях $C$ выполняется условие устойчивости $|q|\le 1$ для заданной гармоники $kh$, можно использовать решатель неравенств [`solve_univariate_inequality`](https://docs.sympy.org/latest/modules/solvers/inequalities.html):

```python
abs_q = abs_q.subs(dx,1/k)                 # зададим гармонику kh = 1
solve_univariate_inequality(abs_q<=1,C)
```

&nbsp; Как мы видим, условие $|q|<1$ выполняется при $C \le 1$ для гармоники с приведенным волновым числом $k \Delta x = 1$. Проиллюстрируем зависимость $|q(C, k \Delta x) |$ графически с помощью функции `q_func_upwind(C,k,dx)`:

```python
import numpy as np
import matplotlib.pyplot as plt

dx = 0.005                      # нам важна комбинация kdx, так что конкретная величина dx не имеет значения
k = 1/dx                        # kdx = 1
C = np.linspace(0, 2, 100)      # С изменяется в пределах от 0 до 2
q = q_func_upwind(C,k,dx)       # |q| рассчиывается с помощью функции q_func_upwind для схемы уголок

fig = plt.figure(figsize=(7, 6))
ax = plt.axes()

font = {'family': 'serif',
        'fontname': 'Alegreya Sans',
        'color':  'black',
        'weight': 'bold',
        'size': 15,
        'fontstyle': 'italic'
        }

ax.set_title('$q(C, k \Delta x = 1)$ для схемы уголок', fontsize = 30,fontdict=font)

ax.set_xlabel("C",fontsize = 30,fontdict=font)
ax.set_ylabel("q", rotation=0,fontsize = 30,fontdict=font)
plt.tight_layout()

plt.text(1.8, 1.3, "q(C)", size=30, rotation=35., fontdict = font,
         ha="center", va="center",
         bbox=dict(boxstyle="round", ec="gray", fc="salmon")
         )
plt.text(0.5, 0.5, "q(C) < 1", size=30,  ha="center", va="center", fontdict = font )
plt.text(0.5, 1.15, "q = 1", size=30,  ha="center", va="center", fontdict = font, color= 'gray' )

ax.fill_between(C, 0, q, where=q <= 1, alpha=0.5, color='green')
ax.fill_between(C, 0, 1, alpha=0.25, color='gray')
ax.plot(C,q,color='black', lw=4)
ax.axhline(1, 0.05, 0.95,ls='--', color='gray', lw=2, alpha=0.5)
```

&nbsp; Анализ графика также показывает, что область устойчивости схемы уголок $C \le 1$ для гармоники $k \Delta x = 1$. Но так ли это для остальных гармоник ? Наиболее наглядно было бы построить поверхность $|q(C, k \Delta x) |$ для всех возможных значений $k\Delta x \in (-\pi, \pi) $. Разработаем для этого функцию `plot_surface(q_func,region,name,dissipation)`:

```python
def plot_surface(q_func,region,name,dissipation):

    '''
        Функция plot_surface служит для построения трехмерных поверхностей |q(C, kdx) | и gamma(C,k\Delta x).
        q_func - функция |q(C, kdx) | или gamma(C,k\Delta x)
        region - область изменения C, например (0,1)
        name - название схемы, например "уголок"
        dissipation - логическая переменная для выбора диссипативной (True) или дисперсионной поверхности (False)
    
    '''
    
    from mpl_toolkits import mplot3d
    import numpy as np
    import matplotlib.pyplot as plt
    
    fig = plt.figure(figsize=(9, 6))
    #ax = plt.axes(projection='3d')
    ax = fig.gca(projection='3d', proj_type = 'ortho')

    font = {'family': 'serif',
        'fontname': 'Alegreya Sans',
        'color':  'black',
        'weight': 'bold',
        'size': 25,
        'fontstyle': 'italic'
        }
    
    if dissipation:
        surf_type = '$|q(C,k \Delta x)|$'
        z_ax = '|q|'
    else:
        surf_type = '$\gamma(C,k\Delta x)$'
        z_ax = 'gamma'
        
    ax.set_title(surf_type+' для схемы '+name+'\n',fontsize = 30, fontdict = font)
        
    ax.set_xlabel("\nC", fontdict = font)
    ax.set_ylabel("\nkdx", rotation=0, fontdict = font)
    ax.set_zlabel(z_ax, rotation=0, fontdict = font)
    ax.view_init(40,-135)

    ax.grid(False)

    dx = 1
    C = np.linspace(region[0], region[1], 100)
    k = np.linspace(-np.pi/dx, np.pi/dx, 100)

    C, K = np.meshgrid(C, k)
    Q1 = np.empty_like(C)
    Q2 = np.empty_like(C)
    C1 = np.empty_like(C, dtype=object)
    C2 = np.empty_like(C, dtype=object)

    for i in range(len(C)):
        for j in range(len(C[0])):
            q1 = q_func(C[i,j],K[i,j],dx)
            q2 = 1.0
            Q1[i,j] = q1
            Q2[i,j] = q2

            if (q1 < 1.001):
                C1[i,j] = plt.get_cmap("gist_stern_r")(q1)
            else:
                C1[i,j] = plt.get_cmap("jet_r")(0.25)
            C2[i,j] = plt.get_cmap("Greys")(q2)
            lst = list(C2[i,j])
            lst[3] = 0.25
            C2[i,j] = tuple(lst)       

    C_bridge = np.vstack([C[-1,:],C[-1,:]])
    K_bridge = np.vstack([K[-1,:],K[-1,:]])
    Q_bridge = np.vstack([Q1[-1,:],Q2[-1,:]])
    color_bridge = np.empty_like(Q_bridge, dtype=object)

    color_bridge.fill((1,1,1,0)) 

    C_full = np.vstack([C, C_bridge, np.flipud(C)])
    K_full = np.vstack([K, K_bridge, np.flipud(K)])
    Q_full = np.vstack([Q1, Q_bridge, np.flipud(Q2)])
    color_full = np.vstack([C1, color_bridge, np.flipud(C2)])

    surf = ax.plot_surface(C_full, K_full, Q_full, rstride=1, cstride=1, 
                           edgecolor='Red', facecolors=color_full, antialiased=True, 
                           cmap='gist_stern_r',linewidth=0.0)
    
    cbar = fig.colorbar(surf, shrink=0.5, aspect=5)
    cbar.set_label(surf_type, labelpad=-25, y=1.25, rotation=0, 
                   fontsize = 20, fontdict = font)
 #   cset = ax.contourf(C, K, Q1, levels =100, zdir='z', offset=0, cmap='Greys')
```

&nbsp; Используем функцию `plot_surface` для построения поверхности $|q(C, k \Delta x) |$ для схемы "уголок" в пределах изменения $C \in (0,1.1)$. 

```python
plot_surface(q_func_upwind,(0.0001,1.1),'уголок',dissipation=True) # в некоторых случаях вблизи 0 возникают 
                                                                   # сингулярности, поэтому делаем небольшой 
                                                                   # отступ
```

&nbsp; Из рисунка мы видим, что значение $|q(C, k \Delta x) |$ меньше 1 для всех гармоник $k\Delta x \in (-\pi, \pi) $ при числе $C \in (0,1.0]$. Однако при $C > 1.0$, множитель перехода $|q(C, k \Delta x) |$ начинает превышать 1 для всех гармоник $k\Delta x \ne 1$, на рисунке область поверхности, где значение $|q(C, k \Delta x) | > 1$ закрашена оранжевым. Таким образом схема уголок устойчива в диапазоне $C \in (0,1.0]$.  

&nbsp; Построим теперь дисперсионную поверхность $|\gamma(C, k \Delta x) |$ для схемы уголок:

```python
plot_surface(g_func_upwind,(0.0001,1.1),'уголок',dissipation=False)
```

&nbsp; График показывает, что схема уголок обладает нормальной дисперсией ($|\gamma(C, k \Delta x) | < 1$) при $C \in (0,0.5]$. Соответственно, при $C > 0.5$ наблюдается аномальная дисперсия. 

&nbsp; Проведем аналогичный анализ для схем Лакса-Вендроффа и Бима-Уорминга: 

```python
q, abs_q, C, k, dx = symbols("q, abs_q, C,k,dx", real=True)

q = C*(C*cos(k*dx)+I*sin(k*dx))-(C**2-1)
abs_q = (Abs(q)).rewrite(cos).simplify().collect(cos(dx*k))
arg_q = Abs(-I/k/dx/C*arg(q))
q_func_lax_wendroff = lambdify([C,dx,k],abs_q)
g_func_lax_wendroff = lambdify([C,dx,k],arg_q)

q = 1.0-0.5*C*(exp(2*I*k*dx)-4*exp(I*k*dx)+3) + 0.5*C**2*(exp(I*k*dx)-1)**2
abs_q = (Abs(q)).rewrite(cos).simplify().collect(cos(dx*k))
arg_q = Abs(-I/k/dx/C*arg(q))
q_func_beam_warming = lambdify([C,dx,k],abs_q)
g_func_beam_warming = lambdify([C,dx,k],arg_q)
```

&nbsp; Построим диссипативные и дисперсионые поверхности схем Лакса-Вендроффа и Бима-Уорминга:

```python
plot_surface(q_func_lax_wendroff,(0.0001,1.1),'Лакса-Вендроффа',dissipation=True)
```

```python
plot_surface(g_func_lax_wendroff,(0.0001,1.1),'Лакса-Вендроффа',dissipation=False)
```

```python
plot_surface(q_func_beam_warming,(0.0001,2.1),'Бима-Уорминга',dissipation=True)
```

```python
plot_surface(g_func_beam_warming,(0.0001,2.1),'Бима-Уорминга',dissipation=False)
```

&nbsp; Из анализа полученных графиков, можно видеть, что диссипативная и дисперсионная поверхности схемы Лакса-Вендроффа ближе прилегают к плоскости $q=1$, чем поверхности схемы уголок, что говорит о более высокой аппороксимации схемы Лакса-Вендроффа. При этом области устойчивости у обоих схем совпадают: $C \le 1$. Схема Уорминга-Бима устойчива при $C \le 2$. При $C = 1$ и $C = 2$ наблюдается интересная особенность - $|q(C, k \Delta x) |$ и $|\gamma(C, k \Delta x) |$ становятся равны 1.0, что говорит о том, что мы получаем точное решение, волновые гармоники численного решения не диссипируют и не диспергируют. Значения $C$, при которых точность решения становится аномально высокой называются *каналами точности* численной схемы. Таким образом, у схемы Бима-Уорминга два канала точности, у схемы уголок и схемы Лакса-Вендроффа - один канал точности на границе области устойчивости, при $C=1$.

#### Задача 9
> Вывести характеристическое уравнение схемы FTCS и построить для этой схемы диссипативную и дисперсионную поверхности.

---

&nbsp; Для трехслойных схем, ввиду того что начальные условия необходимо задавать не на одном а на двух первых временных слоях, характеристическое уравнение принимает вид квадратного трехчлена. К примеру для схемы "крест", характеристическое уравнение выглядит следующим образом:

\begin{equation}
q^2 - 2i q C \sin(k \Delta x) -1  = 0
\label{eq:charact_leap_frog}
\end{equation}

данное уравнение имеет два корня:

\begin{equation}
q_{1,2} = -i C \sin(k \Delta x) \pm \sqrt{1-C^2\sin^2(k\Delta x)}
\label{eq:charact_leap_frog_sol}
\end{equation}

&nbsp; Один из корней описывает эволюцию во времени начального рассогласования данных на первых двух временных слоях. Мы будем называть этот корень "паразитным". При использовании трехслойных схем, решение на втором временном слое (на первом ставится начальное условие) ищется как правило с помощью какой-нибудь двухслойной схемы, и только после того как решение на втором слое найдено, становится возможным использование трехслойных схем. Поэтому рассогласование между первыми двумя слоями определяется свойствами двухслойной схемы. Если модуль паразитного корня меньше 1, ошибка согласования со временем затухает. Второй корень мы будем называть "реальным", именно он отражает насколько дисперсионные и диссипативные свойства схемы отражают свойства исходного уравнения. Для того чтобы различить какой из корней является паразитным, а какой - реальным, необходимо построить для них дисперсионную и диссипативную поверхности. Дисперсионная и диссипативная поверхности реального корня должны бить ближе к единичной поверхности при стремлении волнового числа к нулю при всех числах Куранта из области устойчивости, так как именно реальный корень определяет решение. Для устойчивости схемы модуль обоих корней должен быть не выше 1. 

&nbsp; Диссипативная и дисперсионная поверхности для корня $q_{1} = -i C \sin(k \Delta x) + \sqrt{1-C^2\sin^2(k\Delta x)}$:

```python
q1, abs_q, C, k, dx = symbols("q1, abs_q, C,k,dx", real=True)

q1 = -I*C*sin(k*dx)+sqrt(1-C**2*sin(k*dx)**2)
abs_q = (Abs(q1)).rewrite(cos).simplify().collect(cos(dx*k))
arg_q = Abs(-I/k/dx/C*arg(q1))
q1_func_leap_frog = lambdify([C,dx,k],abs_q)
g1_func_leap_frog = lambdify([C,dx,k],arg_q)
```

```python
plot_surface(q1_func_leap_frog,(0.0001,1.0),'крест ($q_1$)',dissipation=True)
```

```python
plot_surface(g1_func_leap_frog,(0.0001,1.0),'крест ($q_1$)',dissipation=False)
```

&nbsp; Диссипативная и дисперсионная поверхности для корня $q_{2} = -i C \sin(k \Delta x) - \sqrt{1-C^2\sin^2(k\Delta x)}$:

```python
q2, abs_q, C, k, dx = symbols("q2, abs_q, C,k,dx", real=True)

q2 = -I*C*sin(k*dx)-sqrt(1-C**2*sin(k*dx)**2)
abs_q = (Abs(q2)).rewrite(cos).simplify().collect(cos(dx*k))
arg_q = Abs(-I/k/dx/C*arg(q2))
q2_func_leap_frog = lambdify([C,dx,k],abs_q)
g2_func_leap_frog = lambdify([C,dx,k],arg_q)
```

```python
plot_surface(q2_func_leap_frog,(0.0001,1.0),'крест ($q_2$)',dissipation=True)
```

```python
plot_surface(g2_func_leap_frog,(0.0001,1.0),'крест ($q_2$)',dissipation=False)
```

&nbsp; Из анализа диссипативных и дисперсионных поверхностей для корней характеристического уравнения схемы "крест" мы видим, что корень $q_1$ является реальным, т.к. его дисперсионная поверхность ближе прилегает к единичной плоскости при $k\Delta x = 0$, чем поверхность для корня $q_2$. Также необходимо отметить диссипативные свойства схемы "крест". Для любых значений $C \le 1$ и $k\Delta x \in (-\pi,\pi)$ $|q(C, k \Delta x) | = 1$, т.е. схема *бездиссипативна*. При $C > 1$ схема теряет устойчивость, что можно заметить исходя из величины подкоренного выражения $\sqrt{1-C^2\sin^2(k\Delta x)}$ в обоих корнях $q_{1,2}$.
